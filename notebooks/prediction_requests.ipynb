```python
import json
import requests
import pandas as pd

from sklearn.preprocessing import LabelEncoder
```

```python
url = 'https://ml-api-hhcnswtjha-uc.a.run.app/predict'
```

```python
headers = {
    'Authorization': 'Basic YWRtaW46MTIzNDU2',
    'Content-Type': 'application/json'
}
```

```python
# $ gsutil -m cp -r gs://artefatos-do-modelo-dsjobs/ds-job-test.csv dataset/

df_teste = pd.read_csv('dataset/ds-job-test.csv')
df_teste.rename(columns={'enrollee_id':'id'}, inplace=True)
df_teste.head()
```

```python
le = LabelEncoder()
df_teste.loc[:, 'city'] = le.fit_transform(df_teste.loc[:, 'city'])
```

```python
dict_requisitions = df_teste.to_dict(orient='index')
for i, test_example in enumerate(dict_requisitions.values()):
    payload = json.dumps(test_example)
    response = requests.request("POST", url, headers=headers, data=payload)
    print(i+1, response.text)
```
